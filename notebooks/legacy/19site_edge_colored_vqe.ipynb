# 19-Site Edge-Colored HVA: Continuous VQE Pilot

This notebook runs a continuous Powell optimization over the 4-color HVA parameters using Qiskit `Statevector` expectation values. The default budget is intentionally modest; increase `--maxfev`, `--restarts`, and `--max-layers` for a serious optimization campaign.

```python
from pathlib import Path
import subprocess
import sys

import pandas as pd

PROJECT = Path.cwd()
if PROJECT.name == "notebooks":
    PROJECT = PROJECT.parent
```

## Run Continuous Optimization

```python
cmd = [
    sys.executable,
    str(PROJECT / "scripts" / "legacy" / "run_19site_edge_colored_vqe.py"),
    "--max-layers",
    "1",
    "--maxfev",
    "20",
    "--restarts",
    "1",
]
result = subprocess.run(cmd, cwd=PROJECT, text=True, capture_output=True, check=True)
print(result.stdout)
```
